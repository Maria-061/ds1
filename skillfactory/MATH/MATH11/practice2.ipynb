```python
import pandas as pd
```

```python
df = pd.read_csv('data/Country-data.csv')
df['life_expec'].max()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['country'], axis=1)
y = df['country']
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
```

```python
X_scaler[0][0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_df = pd.DataFrame(X_scaler, columns=X.columns)
plt.figure(figsize = (20,10))
sns.heatmap(X_df.corr(), annot=True)
```

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components=0.9)
pca.fit(X_scaler)
len(pca.components_)
```

```python
pca.explained_variance_ratio_[0]
```

```python
import numpy as np
```

```python
X_pca = pca.fit_transform(X_scaler)
corrmat = np.corrcoef(X_pca.transpose())
sns.heatmap(corrmat, annot=True)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
for i in range(2, 11):
    model = KMeans(n_clusters=i, random_state=1)
    model.fit(X_pca)
    print (i, silhouette_score(X_pca, model.labels_))
```

```python
best_model = KMeans(n_clusters=4, random_state=1)
best_model.fit(X_pca)
```

```python
df['K_Means_labels']= best_model.labels_
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.boxplot(x='K_Means_labels', y='gdpp', data=df)
plt.figure(figsize=(8,4),dpi=200)
sns.boxplot(x='K_Means_labels', y='child_mort', data=df)
```

```python
K_Means_countries = df[df['K_Means_labels'] == 1]
K_Means_countries.sort_values(by = 'life_expec')
```
