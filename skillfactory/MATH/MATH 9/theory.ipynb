```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/wineQualityReds.csv')
df.head()
```

```python
df['is_good']=df['quality'].apply(lambda x: 1 if x>=6 else 0)
```

```python
df = df.drop(['quality'], axis=1)
```

```python
df.info()
```

```python
X = df.drop('is_good', axis=1)
y = df['is_good']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import f1_score
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_train_pred=log_reg.predict(X_train)
y_test_pred=log_reg.predict(X_test)

print (f'Метрика F1 на обучающем наборе {round(f1_score(y_train, y_train_pred), 3)}, на тестовом {round(f1_score(y_test, y_test_pred), 3)}')
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tree = DecisionTreeClassifier(random_state=42, max_depth=10)

dec_tree.fit(X_train, y_train)

y_train_pred1=dec_tree.predict(X_train)
y_test_pred1=dec_tree.predict(X_test)

print (f'Метрика F1 на обучающем наборе {round(f1_score(y_train, y_train_pred1), 3)}, на тестовом {round(f1_score(y_test, y_test_pred1), 3)}')
```
